# Helper Functions

```python
def standardize_train_set(X):
    #todo
    mu = np.mean(X, axis=0) #np.sum(X) / float(len(X))
    #todo
    sigma = np.std(X, axis=0, ddof=1)
    #todo
    X_norm = (X - mu)/sigma
    return X_norm, mu, sigma
```

```python
def standardize_test_set(X, mu, sigma):
    #todo
    return (X - mu)/sigma
```

```python
class LinearRegression:
    def __init__(self, alpha=0.001, iters=1000, verbose=False):
        self.alpha = alpha
        self.iters = iters
        self.verbose = verbose
        self.theta = None
    
    def calculate_absolute_error(self, X, y, theta):
        #todo
        h_theta = np.dot(X, theta)
        err = (1/len(y)) * np.sum(abs(h_theta - y))
        return err
    
    def gradient(self, X, y, theta):
        m = len(y)

        #todo
        h_theta = np.dot(X, theta)
        loss = h_theta - y

        #todo
        dJ = (1/m) * np.dot(X.T, loss)
        J = np.sum((loss ** 2) / (2*m))

        return (J, dJ)
    
    def gradient_descent(self, X, y):
        X = np.c_[np.ones(len(X),dtype='int64'), X]
        theta = np.zeros(X.shape[1])
        print(f'The total of training sample: {len(y)}')
        for i in range(self.iters):
            #todo
            J, dJ = self.gradient(X, y, theta)
            #todo
            theta = theta - self.alpha * dJ
            
            # temp_theta = deepcopy(theta)
            # for j in range(len(temp_theta)):
            #     d = (1/len(y)) * np.sum((np.dot(X, theta) - y) * np.array(X[:,j]).reshape(len(y), 1))
            #     temp_theta[j] = theta[j] - self.alpha * d
            # theta = temp_theta
            
            if self.verbose:
                print(f'Iter {i + 1}, loss = {self.calculate_absolute_error(X, y, theta)}')
        return theta
    
    def fit(self, X, y):
        self.theta = self.gradient_descent(X, y)
    
    def predict(self, X):
        results = []
        for Xi in X:
            #todo
            temp = np.dot(np.insert(Xi,0,1), self.theta)
            results.append(temp)
        return results
```

# Main Function

## Train Model

```python
#Load config
with open('config.json',) as f:
    configs = json.load(f)
#Feature-Label split
#todo
X = (df.loc[:, "size (square feet)":"number of bedrooms"]).to_numpy()
#todo
y = df.values[:, 2]
#Standardize feature and Train model
X, mu, sigma = standardize_train_set(X)
model = LinearRegression(alpha= configs['alpha'], iters=configs['iters'], verbose=False)
model.fit(X, y)
```

## Visualize Model

```python
x_surf, y_surf = np.meshgrid(np.linspace(df['size (square feet)'].min(), df['size (square feet)'].max()),
                 np.linspace(df['number of bedrooms'].min(), df['number of bedrooms'].max()))

surf_df = pd.DataFrame({'size (square feet)': x_surf.ravel(), 'number of bedrooms': y_surf.ravel()})
predicted = np.array(model.predict(standardize_test_set(surf_df.values, mu, sigma)))
```

```python
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df['size (square feet)'], df['number of bedrooms']
           ,df['price ($)'], c='red', marker='o', alpha=0.5)
ax.plot_surface(x_surf, y_surf, predicted.reshape(x_surf.shape), color='b', alpha=0.3)

ax.set_xlabel('size (square feet)')
ax.set_ylabel('number of bedrooms')
ax.set_zlabel('price ($)')
plt.show()
```

- Comment: 3D figure shows that not only the higher size, the higher price but the price also tends to increase with the number of bedrooms

## Predict Sample

```python
sample = [1650,3]
#todo
standardize_sample = standardize_test_set(np.array(sample), mu, sigma)
y_predict = model.predict(np.asmatrix(standardize_sample).tolist())
print(f'House price expected is {round(y_predict[0])}$')
```
